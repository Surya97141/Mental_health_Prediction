```python
summary=[]
for col in df.columns:
    summary.append((
        col,
        df[col].dtype,
        df[col].nunique,
        df[col].isnull().sum(),
        df[col].value_counts(dropna=False).head(3).to_dict()))
    
import pandas as pd
summary_df = pd.DataFrame(summary, columns=['Column', 'Type', 'Unique', 'Nulls', 'Top Responses'])
summary_df.head(10)
```

```python
missing=summary_df[['Column', 'Nulls']]
print(missing.sort_values(by='Nulls', ascending=False).head(10))
```
